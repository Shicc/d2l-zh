```python
# 试试三个类别的小数据
from mxnet.gluon import data as gdata
import os

data_dir = '../../../food3'

imgs = gdata.vision.ImageFolderDataset(data_dir)
```

```python
imgs
```

```python
imgs.__len__()
```

```python
# img, label = 
imgs.__getitem__(2099) # 此函数返回 img, label，2099属于第三类，返回label为2
# 上面语句 == imgs[2099]
```

```python
# 指定RGB三个通道的均值和方差来将图像通道归一化
normalize = gdata.vision.transforms.Normalize(
    [0.485, 0.456, 0.406], [0.229, 0.224, 0.225])

augs = gdata.vision.transforms.Compose([
    gdata.vision.transforms.RandomResizedCrop(224),
    gdata.vision.transforms.RandomFlipLeftRight(),
    gdata.vision.transforms.ToTensor(),
    normalize])

iter = gdata.DataLoader(imgs.transform_first(augs), batch_size=100, shuffle=True)
```

```python
iter
```

```python
iter.__len__()
```

```python
train = []
test = []
sep = int(0.7*iter.__len__())
for i, batch in enumerate(iter):
    if(i<sep):
        train += batch
    else:
        test += batch
```
